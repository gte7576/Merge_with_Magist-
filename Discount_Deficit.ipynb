```python

category_dict = {
    'Stands and Supports': ['stand', 'support'],
    'Keyboards and Mice': ['keyboard', 'keypad', 'mouse'],
    'Devices': ['bracelet', 'wireless remote', 'monitor', 'weight'],
    'Cables and Docks': ['dock', 'cable'],
    'Memory and Storage': ['memory', 'ram', 'card', 'server', ' storage', 'hard drive'],
    'Bags and Cases': ['backpack', 'case', 'cover', 'sleeve'],
    'Laptops and Computers': ['apple macBook pro', 'core', 'laptop', 'computer', 'desktop'],
    'Electronics':['olloclip', 'bluetooth', 'adapter', 'sandisk ', 'microphones', 'robot',
                   'thermometer']
    'Iphone accessories': []
}

def categorize_product(row):
    for category, keywords in category_dict.items():
        if any(keyword in row['name'].lower() or keyword in row['desc'].lower() for keyword in keywords):
            return category
    return 'Other'


products_cl['category'] = products_cl.apply(categorize_product, axis=1)


products_cl
```
